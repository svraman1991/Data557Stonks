```python
import pandas as pd
import numpy as np
```

```python
# Hyperparamters
threshold = 100
percent_profit = 20
percent_SL = 10
# Input Dataset
df = pd.read_csv('RSI_CCI_MACD.csv')
```

```python
def CCI_Analysis(df, threshold, percent_profit, percent_SL):
    cci34 = df[df['CCI'] >= threshold]
    count = len(cci34)
    indices = cci34.index.tolist()

    print(f'Number of times the CCI value is greater than or equal to {threshold}: {count} \n\n')

    for i in range(len(indices)):
        idx = indices[i]
        buy_price = df.loc[idx, 'CLOSE']
        target_price = buy_price + (percent_profit/100) * buy_price
        stop_loss_price = buy_price - (percent_SL/100) * buy_price
        sell_idx = cci34.index.searchsorted(idx+1)
        count_tp_breach = 0
        count_sl_breach = 0
        if i != len(indices) - 1:
            for j in range(i+1, len(indices)):
                tp_idx = indices[j]
                sl_idx = indices[j]
                if df.loc[tp_idx, "CLOSE"] >= target_price:
                    count_tp_breach += 1
                if df.loc[sl_idx, "CLOSE"] <= stop_loss_price:
                    count_sl_breach += 1

            stock_price_cci_less_threshold = None
            for k in range(i+1, len(df)):
                if df.loc[k, "CCI"] < threshold and not np.isnan(df.loc[k, "CCI"]):
                    stock_price_cci_less_threshold = df.loc[k, "CLOSE"]
                    break

            print(f'***************FOR INDEX {idx}***************')
            print(f'Buy Price: {buy_price:.2f}')
            print(f'Target Price: {target_price:.2f}')
            print(f'Stop Loss Price: {stop_loss_price:.2f}')
            print(f'Number of times stock price went above Target Price: {count_tp_breach}')
            print(f'Number of times stock price went below Stop Loss Price: {count_sl_breach}')
            print(f'Stock Price value when CCI became less than {threshold}: {stock_price_cci_less_threshold}')
            #print(f'Profit (in positive) or loss (in negative) had we not traded at target price: {profit_or_loss:.2f}\n\n')
            print("\n")

        else:
            pass
```

```python
CCI_Analysis(df, threshold, percent_profit, percent_SL)
```
