```python
def welch_t_test(df_stock,df_n50):
    '''
    Compare the top 100 stocks with the nifty50 data. Welch_t_test.
    H0: No difference between mean stock returns vs mean nifty50 returns
    HA: Stock return is greater than that of nifty50
    '''
    stock_names = ['ZOMATO', 'WIPRO', 'MCDOWELL-N', 'ULTRACEMCO', 'TORNTPHARM',
       'TITAN', 'TECHM', 'TATASTEEL', 'TATAPOWER', 'TATAMOTORS',
       'TATACONSUM', 'TCS', 'SUNPHARMA', 'SIEMENS', 'SHREECEM',
       'MOTHERSON', 'SRF', 'SBILIFE', 'SBICARD', 'RELIANCE', 'PGHH',
       'POWERGRID', 'PIDILITIND', 'PIIND', 'PAYTM', 'ONGC', 'NESTLEIND',
       'MPHASIS', 'MARUTI', 'MARICO', 'M&M', 'LICI', 'LT', 'LTIM',
       'INDIGO', 'INFY', 'NAUKRI', 'INDUSINDBK', 'INDUSTOWER', 'IRCTC',
       'IOC', 'ITC', 'ICICIPRULI', 'ICICIGI', 'ICICIBANK', 'HINDUNILVR',
       'HAL', 'HINDALCO', 'HEROMOTOCO', 'HAVELLS', 'HDFCLIFE', 'HDFCBANK',
       'HDFCAMC', 'HCLTECH', 'GRASIM', 'GODREJCP', 'GLAND', 'GAIL',
       'NYKAA', 'EICHERMOT', 'DRREDDY', 'DIVISLAB', 'DABUR', 'DLF',
       'COLPAL', 'COALINDIA', 'CIPLA', 'CHOLAFIN', 'BOSCHLTD', 'BIOCON',
       'BHARTIARTL', 'BPCL', 'BEL', 'BERGEPAINT', 'BANKBARODA',
       'BANDHANBNK', 'BAJAJHLDNG', 'BAJAJFINSV', 'BAJFINANCE',
       'BAJAJ-AUTO', 'AXISBANK', 'DMART', 'ASIANPAINT', 'APOLLOHOSP',
       'AMBUJACEM', 'ADANITRANS', 'ATGL', 'ADANIPORTS', 'ADANIGREEN',
       'ADANIENT', 'ACC']
    
    stock_list = []
    counter = 0
    for i in stock_names:
        df_stock_new = df_stock[df_stock['SYMBOL'] == i]
        df_stock_new = df_stock_new.dropna()
        stock_name = i
        print(f'****Results for {stock_name} ****')
    
        # Welch t-test compares nifty50 to individual stock.
        # Alternative Hypothesis is that the indivual stock returns exceed nifty 50 returns
        test_results = scipy.stats.ttest_ind(df_stock_new['Returns'], df_n50['Returns'], equal_var = False, alternative = 'greater')
        print(f'Test Statistic: {test_results.statistic}')
        print(f'p-value: {test_results.pvalue} \n')
    
        if test_results.pvalue < 0.05:
            counter +=1
            stock_list.append(i)
        else:
            pass
    
    
    print(f'Number of stocks that individually surpass nifty50: {counter}')
    print(f'Out of {len(stock_names)} stocks')
    print(stock_list)
    
    return
```

```python
def assumption_test(df_stock):
    '''
    Performs skew test and kurtosis test for all stock returns.
    Skewtest:
    H0: The difference between skewness of the distribution compared to the normal dist. is not significant
    HA: The difference between skewness of the distrbituion compared to the normal dist. is significiant
    
    H0 being rejected implies that we cannot assume the distribution to be normal when considering skewness.
    
    Kurtosistest:
    H0: The difference between kurtosis of the distribution compared to the normal dist. is not significant
    HA: The difference between kurtosis of the distrbituion compared to the normal dist. is significiant
    
    H0 being rejected implies that we cannot assume the distribution to be normal when considering kurtosis.
    
    '''
    stock_names = ['ZOMATO', 'WIPRO', 'MCDOWELL-N', 'ULTRACEMCO', 'TORNTPHARM',
       'TITAN', 'TECHM', 'TATASTEEL', 'TATAPOWER', 'TATAMOTORS',
       'TATACONSUM', 'TCS', 'SUNPHARMA', 'SIEMENS', 'SHREECEM',
       'MOTHERSON', 'SRF', 'SBILIFE', 'SBICARD', 'RELIANCE', 'PGHH',
       'POWERGRID', 'PIDILITIND', 'PIIND', 'PAYTM', 'ONGC', 'NESTLEIND',
       'MPHASIS', 'MARUTI', 'MARICO', 'M&M', 'LICI', 'LT', 'LTIM',
       'INDIGO', 'INFY', 'NAUKRI', 'INDUSINDBK', 'INDUSTOWER', 'IRCTC',
       'IOC', 'ITC', 'ICICIPRULI', 'ICICIGI', 'ICICIBANK', 'HINDUNILVR',
       'HAL', 'HINDALCO', 'HEROMOTOCO', 'HAVELLS', 'HDFCLIFE', 'HDFCBANK',
       'HDFCAMC', 'HCLTECH', 'GRASIM', 'GODREJCP', 'GLAND', 'GAIL',
       'NYKAA', 'EICHERMOT', 'DRREDDY', 'DIVISLAB', 'DABUR', 'DLF',
       'COLPAL', 'COALINDIA', 'CIPLA', 'CHOLAFIN', 'BOSCHLTD', 'BIOCON',
       'BHARTIARTL', 'BPCL', 'BEL', 'BERGEPAINT', 'BANKBARODA',
       'BANDHANBNK', 'BAJAJHLDNG', 'BAJAJFINSV', 'BAJFINANCE',
       'BAJAJ-AUTO', 'AXISBANK', 'DMART', 'ASIANPAINT', 'APOLLOHOSP',
       'AMBUJACEM', 'ADANITRANS', 'ATGL', 'ADANIPORTS', 'ADANIGREEN',
       'ADANIENT', 'ACC']
    
    counter_skew_test = 0
    counter_kurtosis_test = 0
    counter = 0
    skew_not_neglectable = []
    kurtosis_not_neglectable = []
    skew_statistics = []
    kurtosis_stats = []
    pvals_skew = []
    pvals_kurtosis = []
    results_skew = []
    results_kurtosis = []
    
    for i in stock_names:
        df_stock_new = df_stock[df_stock['SYMBOL'] == i]
        df_stock_new = df_stock_new.dropna()
        stock_name = i
        
        if df_stock_new.shape[0] <8:
            print(df_stock_new['SYMBOL'])
            continue
        else:
            counter += 1
        
        print(f'****Results for {stock_name} ****')
    
        # Alternative Hypothesis is that the indivual stock assumption is not valid.
        skew_results = skewtest(df_stock_new['Returns'])
        kurtosis_results = kurtosistest(df_stock_new['Returns'])
        
        print(f'**** Skew Test Results ****')
        print(f'Test Statistic: {skew_results.statistic}')
        print(f'p-value: {skew_results.pvalue} \n')
        
        print(f'**** Kurtosis Test Results ****')
        print(f'Test Statistic: {kurtosis_results.statistic}')
        print(f'p-value: {kurtosis_results.pvalue} \n')
        
        skew_statistics.append(skew_results.statistic)
        kurtosis_stats.append(kurtosis_results.statistic)
        pvals_skew.append(skew_results.pvalue)
        pvals_kurtosis.append(kurtosis_results.pvalue)

        if skew_results.pvalue < 0.05:
            counter_skew_test +=1
            results_skew.append("Reject Null")
        else:
            results_skew.append("Fail to Reject")
    
        
        if kurtosis_results.pvalue < 0.05:
            counter_kurtosis_test +=1
            results_kurtosis.append("Reject Null")
        else:
            results_kurtosis.append("Fail to Reject")
    
    
    df_assumptions = pd.DataFrame({
                                   'skew test-statistic': skew_statistics,
                                   'Skew test p-value': pvals_skew,
                                   'Skew test results': results_skew,
                                   'kurtosis test-statistic': kurtosis_stats,
                                   'Kurtosis test p-value': pvals_skew,
                                   'Kurtosis Test Results': results_kurtosis
                              })
    
    df_assumptions.to_csv('Assumption Results.csv')
    stocks_df = pd.DataFrame({"Stock Name": stock_names})
    
    stocks_df.to_csv('names.csv')
    print(f'Number of stocks that do not satisfy skew test assumptions: {counter_skew_test}')
    print(f'Number of stocks that do not satisfy kurtosis test assumptions: {counter_kurtosis_test}')
    print(f'Number of eligible stocks based on data: {len(stock_names) - counter}')
```

```python
# T Test for niftybee vs other stocks
def t_test_rolling_res(df_stock,df_n50):
    '''
    This is the t-test comparing the stock reutrn means vs the nifty50 rolling returns.
    This does not incorporate the indicator-flag approach. 
    '''
    stock_names = ['ZOMATO', 'WIPRO', 'MCDOWELL-N', 'ULTRACEMCO', 'TORNTPHARM',
       'TITAN', 'TECHM', 'TATASTEEL', 'TATAPOWER', 'TATAMOTORS',
       'TATACONSUM', 'TCS', 'SUNPHARMA', 'SIEMENS', 'SHREECEM',
       'MOTHERSON', 'SRF', 'SBILIFE', 'SBICARD', 'RELIANCE', 'PGHH',
       'POWERGRID', 'PIDILITIND', 'PIIND', 'PAYTM', 'ONGC', 'NESTLEIND',
       'MPHASIS', 'MARUTI', 'MARICO', 'M&M', 'LICI', 'LT', 'LTIM',
       'INDIGO', 'INFY', 'NAUKRI', 'INDUSINDBK', 'INDUSTOWER', 'IRCTC',
       'IOC', 'ITC', 'ICICIPRULI', 'ICICIGI', 'ICICIBANK', 'HINDUNILVR',
       'HAL', 'HINDALCO', 'HEROMOTOCO', 'HAVELLS', 'HDFCLIFE', 'HDFCBANK',
       'HDFCAMC', 'HCLTECH', 'GRASIM', 'GODREJCP', 'GLAND', 'GAIL',
       'NYKAA', 'EICHERMOT', 'DRREDDY', 'DIVISLAB', 'DABUR', 'DLF',
       'COLPAL', 'COALINDIA', 'CIPLA', 'CHOLAFIN', 'BOSCHLTD', 'BIOCON',
       'BHARTIARTL', 'BPCL', 'BEL', 'BERGEPAINT', 'BANKBARODA',
       'BANDHANBNK', 'BAJAJHLDNG', 'BAJAJFINSV', 'BAJFINANCE',
       'BAJAJ-AUTO', 'AXISBANK', 'DMART', 'ASIANPAINT', 'APOLLOHOSP',
       'AMBUJACEM', 'ADANITRANS', 'ATGL', 'ADANIPORTS', 'ADANIGREEN',
       'ADANIENT', 'ACC', 'NIFTYBEES']
    
    stock_list = []
    counter = 0
    pvals = []
    stat = []
    sig = []
    
    df_stock = df_stock.dropna()
    df_n50 = df_n50.dropna()
    
    for i in stock_names:
        df_stock_new = df_stock[df_stock['SYMBOL'] == i]
        df_stock_new = df_stock_new.dropna()
        stock_name = i
        print(f'****Results for {stock_name} ****')
    
        # Welch t-test compares nifty50 to individual stock.
        # Alternative Hypothesis is that the indivual stock returns exceed nifty 50 returns
        stock = df_stock_new['Rolling_1year'].dropna()
        nift = df_n50['Rolling_1year'].dropna()
        test_results1 = scipy.stats.ttest_ind(stock, nift, equal_var = False, alternative = 'greater')
        print(f'Test Statistic: {test_results1.statistic}')
        print(f'p-value: {test_results1.pvalue} \n')
        
        pvals.append(test_results1.pvalue)
        stat.append(test_results1.statistic)
    
        if test_results1.pvalue < 0.05:
            sig.append('Reject Null')
            counter +=1
            stock_list.append(i)
        else:
            sig.append('Fail to Reject Null')
    
    df_results = pd.DataFrame({'Stock Name': stock_names,
                               'test-statistic':stat,
                              'p-value': pvals,
                              'Conclusion':sig})
    print(f'Number of stocks that individually surpass nifty50: {counter}')
    print(f'Out of {len(stock_names)} stocks')
    print(stock_list)
    
    return (df_results)
```
